```python
from optimum.intel import OVModelForCausalLM, OVWeightQuantizationConfig
```

```python
model_name="HannahJohn/finetuned_llama2-7b-chat-hf"
output_dir="openvino"
```

```python
quantization_config = OVWeightQuantizationConfig(bits=4,sym='True')
```

```python
ov_model = OVModelForCausalLM.from_pretrained(model_name, export=True, quantization_config=quantization_config)
```

```python
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
output_dir="/content/results"
```

```python
ov_model.save_pretrained(output_dir)
tokenizer.save_pretrained(output_dir)
```
